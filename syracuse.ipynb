```python
# Suite de Syracuse
from matplotlib.pyplot import *
x=int(input("Position de départ: "))
L=[x]
while (x!=1):
    if x%2==0:
        x=x/2
    else:
        x=3*x+1
    L=L+[x]
xlabel("temps de vol")
ylabel("altitude")
plot(L)
print("Temps de vol: ",len(L))
print("Altitude maximale: ", int(max(L)))
```
